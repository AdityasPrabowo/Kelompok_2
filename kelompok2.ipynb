```python
class Pengguna:
    def __init__(self, username: str, email: str):
        self.username = username
        self.email = email
        self.status_aktif = False

    def aktifkan_akun(self):
        if not self.status_aktif:
            self.status_aktif = True
            print(f"Akun {self.username} telah diaktifkan.")
        else:
            raise ValueError(f"Akun {self.username} sudah aktif.")

    def nonaktifkan_akun(self):
        if self.status_aktif:
            self.status_aktif = False
            print(f"Akun {self.username} telah dinonaktifkan.")
        else:
            raise ValueError(f"Akun {self.username} sudah nonaktif.")

daftar_pengguna = []
daftar_pengguna_dict = {}

def tampilkan_semua_pengguna():
    if not daftar_pengguna:
        print("Tidak ada pengguna yang terdaftar.")
        return
    for pengguna in daftar_pengguna:
        status = "Aktif" if pengguna.status_aktif else "Nonaktif"
        print(f"Username: {pengguna.username}, Email: {pengguna.email}, Status: {status}")

def cari_pengguna(username: str):
    pengguna = daftar_pengguna_dict.get(username)
    if pengguna:
        status = "Aktif" if pengguna.status_aktif else "Nonaktif"
        print(f"Username: {pengguna.username}, Email: {pengguna.email}, Status: {status}")
    else:
        print(f"Pengguna dengan username '{username}' tidak ditemukan.")

def tambah_pengguna(username: str, email: str):
    if username in daftar_pengguna_dict:
        print(f"Pengguna dengan username '{username}' sudah ada.")
    else:
        pengguna_baru = Pengguna(username, email)
        daftar_pengguna.append(pengguna_baru)
        daftar_pengguna_dict[username] = pengguna_baru
        print(f"Pengguna '{username}' berhasil ditambahkan.")

def ubah_status_akun(username: str, status: str):
    pengguna = daftar_pengguna_dict.get(username)
    if pengguna:
        try:
            if status.lower() == "aktif":
                pengguna.aktifkan_akun()
            elif status.lower() == "nonaktif":
                pengguna.nonaktifkan_akun()
            else:
                print("Status tidak valid. Gunakan 'aktif' atau 'nonaktif'.")
        except ValueError as e:
            print(e)
    else:
        print(f"Pengguna dengan username '{username}' tidak ditemukan.")

if __name__ == "__main__":
    tambah_pengguna("user1", "kelompokdua@gmail.com")
    tambah_pengguna("user2", "kelompok2@gmail.com")

    print("\nTampilkan semua pengguna setelah penambahan:")
    tampilkan_semua_pengguna()

    ubah_status_akun("user1", "aktif")
    ubah_status_akun("user2", "nonaktif")

    print("\nTampilkan semua pengguna setelah perubahan status:")
    tampilkan_semua_pengguna()

    print("\nCari pengguna:")
    cari_pengguna("user1")
```
